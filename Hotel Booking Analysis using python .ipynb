<b>QUESTION 2. What's the difference between "reserved_room_type" and "assigned_room_type"? and did hotels fulfill the customers expectations? if "yes" or "no" ,how can we improve? </b>

````python
#counting assigned room types
assigned_room_type=df['assigned_room_type'].value_counts()
assigned_room_type=pd.DataFrame(assigned_room_type)
assigned_room_type
````

````python
#creating a list of assigned_room_types bookings
assigned_values=[i for i in df['assigned_room_type'].value_counts()]
print(assigned_values)

#creating a dictionary of assigned_room_types
assigned_rooms_dict=dict(zip(assigned_room_type.index,assigned_values))
print(assigned_rooms_dict)
````

````python
#creating a list of reserved_room_types bookings
reserved_values=[i for i in df['reserved_room_type'].value_counts()]
print(reserved_values)

#creating a dictionary of reserved_room_types
reserved_rooms_dict=dict(zip(reserved_room_type.index,reserved_values))
print(reserved_rooms_dict)
````

````python
#creating a list with reserved_rooms_dict and assigned_rooms_dict
rooms_list=[reserved_rooms_dict,assigned_rooms_dict]
print(rooms_list)

#creating a dataframe with reserved_rooms_dict and assigned_rooms_dict as columns
rooms_df=pd.DataFrame(rooms_list,index=['Reserved rooms','Assigned rooms'])
rooms_df
````

````python
#visualization of diff_list
rooms_df.T.plot(kind='bar',figsize=(15,5))
plt.title('Reserved rooms and assigned rooms')
plt.xlabel('Room types')
plt.ylabel('Number of rooms')
plt.show()
````

````python
#creating a dataframe with a new column that contains difference between reserved and assigned rooms count
difference_df=rooms_df.T
difference_df['Difference']=difference_df['Reserved rooms']-difference_df['Assigned rooms']
difference_df.T
````

````python
#visualization of difference between reserved and assigned rooms count
difference_df['Difference'].plot(kind='bar',figsize=(15,5),color='brown')
plt.title('Difference between count of Reserved rooms and assigned rooms')
plt.xlabel('Room types')
plt.ylabel('Difference')
plt.show()
````

<b>Most of the hotels have met the customers expectations in terms of assigning reserved room types. But should also try to give their maximum to fulfill customers reservations.</b>

<b>QUESTION 5. Which months have highest no of cancellation?</b>

````python
#creating a dataframe containing monthly cancellations records
cancellation_records=df[df['is_canceled']==1]
cancellation_records
cancellations_df=pd.DataFrame(cancellation_records['arrival_date_month'].value_counts()).T
cancellations_df
````

````python
print('cancellations were high in',cancellations_df.columns[0],'with a value of',cancellations_df.max()[0])
````

````python
#creating a list of months
months=['January','February','March','April','May','June','July','August','September','October','November','December']

#creating a list of month wise number of bookings
bookings=[monthly_count[i][0] for i in months]

#creating a list of month wise number of cancellations
cancellations=[cancellations_df[i][0] for i in months]

# creating dictionaries of bookings and cancellations
first_bookings=dict(zip(months,bookings))
successful_cancellations=dict(zip(months,cancellations))

#creating a list of bookings and cancellations
final_bookings=[first_bookings,successful_cancellations]
final_bookings
````

````python
#finding final successful bookings
successful_bookings=pd.DataFrame(final_bookings,index=['Bookings','Cancellations']).T
successful_bookings['Successful_bookings']=successful_bookings['Bookings']-successful_bookings['Cancellations']
successful_bookings.T
````
